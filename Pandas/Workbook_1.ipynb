### Creating a DataFrame

Using Zip 

```python
Names=['Ajay','Sumit','Vinod','Binod']
Age=[23,12,45]

lst=list(zip(Names,Age))
lst

df=pd.DataFrame(lst)
df
```

```python
Names=['Ajay','Sumit','Vinod','Binod']
Age=[23,12,45,32]

lst=list(zip(Names,Age))
lst

df=pd.DataFrame(lst,columns=['Name','Age'])
df
```

### Drop Columns 

```python
df4=pd.read_csv('IMDB-Movie-Data.csv')
df4.head()
```

```python
#first lets make a copy 
df5=df4.copy()
df5
```

```python
#using drop function 
df5.drop('Rank',axis=1,inplace=True)
df5.head()
```

```python
df5.drop(['Title','Genre'],axis=1,inplace=True)
df5.head()
```

```python
#using index value of columns 
df5.drop(df5.columns[[0,1]],axis=1,inplace=True)
df5.head()
```

```python
#using iloc 
df5.drop(df5.iloc[:,0:1],axis=1)
```

```python
df5.head()
```

```python
#using loc 
df5.drop(df5.loc[:,'Actors':'Rating'].columns,axis=1)
```

```python
# using del and for loop 
col=['Rating','Rank','Actors']
for cols in df4.columns:
    if cols in col :
        del df4[cols]
        
df4
```

### Handling Missing Value 

```python
df=pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.shape
```

```python
#finding the null values column-wise
df.isna().sum()
```

```python
#total no of null values 
df.isna().sum().sum()
```

```python
#fillna
df11=df.fillna(value=0)
df11
```

```python
df11.isna().sum()
```

```python
#fill null value with previous value 
df12=df.fillna(method='pad')
df12.head()
```

```python
#fill null value with next value (backward fill)
df13=df.fillna(method='bfill')
df13.head()
```

```python
#fill null value column wise 
df14=df.fillna(method='pad',axis=1)
df14.head()
```

```python
#fill null value column wise ( back filling) 
df15=df.fillna(method='bfill',axis=1)
df15.head()
```

```python
#fill different values in different columns 
df16=df.fillna({'society':'Unknown','balcony':'unavailable'})
df
```

```python
df16
```

```python
#fill null value with mean of numeric column 
df17=df.fillna(value=df['balcony'].mean())
df17
```

```python
df18=df.fillna(value=df['balcony'].max())
df18
```

### from above plot , the employees are classified into ( Fresher , 5+ year exp, 10+ year exp, 20+ year exp & Higher management)

```python
def experience_filter(x):
    if x>=0 and x<5:
        return 'Fresher'
    elif x >= 5 and x<=10:
        return '5+ Year Experience'
    elif x>10 and x<=20:
        return '10+ Year Experience'
    elif x>20 and x<=30:
        return '20+ Year Experience'
    elif x>30:
        return 'Management'
```

### lets have a look at Distance from home to Office

```python
sns.boxplot(data=emp_still_here,x='DistanceFromHome')
```

### from the above plot , we can classify the data into three labels ( Near, Far, Very Far)

```python
def distance_filter(x):
    if x>0 and x<=4:
        return 'Near'
    elif x>4 and x<=10:
        return 'Far'
    elif x>10:
        return 'Very far'
    
```

## lets check the trend of Salary 

```python
grp3=female_emp_data.groupby('Experience')['MonthlyIncome'].agg(['max','min']).sort_values(by='max',ascending=False)
grp3
```

### It is Understood that freshers are paid lowest salary in the whole Organization & Employees at Management Level are paid Highest 

```python
fig, ax = plt.subplots(figsize=(15,7))
grp5=female_emp_data.groupby(['Department','JobRole']).agg({'PercentSalaryHike':'max'})
grp5.plot(kind='barh',ax=ax)
plt.title('Female Employees')
plt.xlabel('Maximum Salary Hike')
```

###  R & D department Employees(Female) have highest Salary Hike 

```python
fig, ax1 =plt.subplots()
grp6=male_emp_data.groupby(['Department','JobRole']).agg({'PercentSalaryHike':'mean'})
grp6.plot(kind='barh',ax=ax1,figsize=(10,8))
plt.title('Male Employees')
plt.xlabel('Average Salary Hike')
```

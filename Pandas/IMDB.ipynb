#  IMDB DataSet

Here's a data set of 1,000 most popular movies on IMDB in the last 10 years(2006 to 2016).

The data points included are: Title, Genre, Description, Director, Actors, Year, Runtime, Rating, Votes, Revenue, Metascrore. 

Link : https://www.kaggle.com/PromptCloudHQ/imdb-data

####  Lets load the IMDB Dataset

```python
dataimdb=pd.read_csv('IMDB-Movie-Data.csv')
```

```python
# sneak-peek of how the Data looks like
dataimdb.head()
```

```python
# Getting the no of Rows , Columns along with their Data Type 
dataimdb.info()
```

```python
# Changing Data-Type Of a column fro integer to float
dataimdb['Runtime (Minutes)']=dataimdb['Runtime (Minutes)'].astype('float')
```

```python
# re-check to see , if the change has been made 
dataimdb.info()
```

```python
# converting Runtime from Minutes to Hours
dataimdb['Runtime (Minutes)']=dataimdb['Runtime (Minutes)']/60
dataimdb.head()
```

```python
# rounding off the values after the decimal 
dataimdb['Runtime (Minutes)']=dataimdb['Runtime (Minutes)'].round(2)
dataimdb
```

```python
# Renaming of a column 
dataimdb.rename(columns={'Runtime (Minutes)':'Runtime (Hours)'},inplace=True)
```

```python
# confirm the changes have been made 
dataimdb.head()
```

```python
# Dropping not Required columns
dataimdb.drop(['Rank','Description','Actors'],axis=1,inplace=True)
```

```python
dataimdb.head()
```

```python
# lets find out , are there any NULL values in DataFrame 
dataimdb.isna().sum()
```

```python
# Get no of rows & columns 
dataimdb.shape
```

```python
# lets drop the NULL values 
dataimdb.dropna(axis=0,how='any',inplace=True)
```

```python
# check the updated rows & columns 
dataimdb.shape
```

```python
# check again if there are any remaining NULL values
dataimdb.isna().sum()
```

```python
# columns of the Dataframe
dataimdb.columns.to_list
```

```python
# split the Dataset with Revenue column in focus , for Financial Analysis
data_revenue=dataimdb[['Year','Genre','Title','Runtime (Hours)','Revenue (Millions)']]
data_revenue.head()
```

```python
# split the Dataset with "Rating" column in focus , for Critics Analysis 
data_rating=dataimdb[['Year','Genre','Title','Director','Rating','Votes','Metascore']]
data_rating.head()
```

```python
# Sorting the Dataset by Year & "Revenue (max to min) "
data_revenue.sort_values(['Year','Revenue (Millions)'],ascending=False,inplace=True)
```

```python
data_revenue.tail()
```

```python
# Sort the Dataset by Year & "Rating( Highest to Minimum )"
data_rating.sort_values(['Year','Rating'],ascending=False,inplace=True)
```

```python
data_rating.head()
```

```python
# Generate a Dataset with Genres & its count 
genres=data_rating['Genre'].value_counts().to_frame()
```

```python
genres.head(20)
```

### Explore Dataset with Visualization

```python
#set the background Style 
sns.set_style('darkgrid')
```

```python
# Maximum Revenue per Year
plt.figure(figsize=(10,6))
sns.lineplot(x=data_revenue['Year'],y=data_revenue['Revenue (Millions)'],estimator='max')
plt.title('Maximum Revenue for Each Year')
```

```python
# Minimun Revenue per Year
plt.figure(figsize=(10,6))
sns.lineplot(x=data_revenue['Year'],y=data_revenue['Revenue (Millions)'],estimator='min')
plt.title('Minimum Revenue of each year')
```

```python
# Mean Revenue per Year
plt.figure(figsize=(10,6))
sns.lineplot(x=data_revenue['Year'],y=data_revenue['Revenue (Millions)'],estimator='mean')
plt.title('Average Revenue of Each Year')
```

```python
# Minimum Metascore per Year
plt.figure(figsize=(10,6))
sns.lineplot(data=data_rating,x='Year',y='Metascore',estimator='min')
plt.title('Lowest Metascore of Each Year')
```

```python
# Maximum Metascore per Year
plt.figure(figsize=(10,6))
sns.lineplot(data=data_rating,x='Year',y='Metascore',estimator='max')
plt.title('Highest Metascore Of Each Year ')
```

```python
# Maximum Runtime of Each Year 
plt.figure(figsize=(10,8))
sns.barplot(data=data_revenue,x='Year',y='Runtime (Hours)',saturation=1,palette="Blues_d",estimator=max)
plt.title('Longest Movies(Runtime) Of each Year')
```

```python
# Minimum Runtime of Each Year 
plt.figure(figsize=(10,8))
sns.barplot(data=data_revenue,x='Year',y='Runtime (Hours)',saturation=0.2,palette="Blues_d",estimator=min)
plt.title('Shortest Movie(Runtime) of each Year')
```

```python
# top 10 revenue generated
top_rev=data_revenue.nlargest(10,'Revenue (Millions)')
plt.figure(figsize=(25,10))
sns.set_style('darkgrid')
sns.lineplot(data=top_rev,x='Title',y='Revenue (Millions)')
plt.title('10 Highest revenue Movies (2006 to 2016) ')
```

```python
# top 10 movies with longest runtime
top_runtime=data_revenue.nlargest(10,'Runtime (Hours)')
plt.figure(figsize=(10,10))
sns.barplot(data=top_runtime,x='Runtime (Hours)',y='Title',orient='h',palette="Blues_d",saturation=1)
plt.title('')
```

```python
# top 10 movies with highest Ratings 
top_rating=data_rating.nlargest(10,'Rating')
plt.figure(figsize=(10,8))
sns.set_style('dark')
sns.barplot(data=top_rating,x='Year',y='Rating',palette="Blues_d",saturation=0.1)
plt.title('Highest Rating (2006 to 2016)')
```

```python
top_rating
```

```python
#top 10 movies with highest votes belong to which genre 
top_votes=data_rating.nlargest(10,'Votes')
plt.figure(figsize=(10,10))
sns.set_style('whitegrid')
sns.barplot(data=top_votes,x='Votes',y='Genre',orient='h')
plt.title('Movies with Highest votes Belonging to Genres')
```

```python
# here we will filter the data_revunue and data_rating Dataset , And check the top 5 movies in terms of Revenue, Rating 
```

```python
cond1=(data_rating['Genre']=='Action,Adventure,Sci-Fi')
df1=data_rating.loc[cond1].nlargest(5,'Rating')
df1.plot(kind='line',x='Title',y='Rating',figsize=(17,10),alpha=0.8,title='Top 5 Highest Rated Movies in Genre(Action,Adventure,Sci-Fi)',color='Green')
plt.xlabel('TITLE')
plt.ylabel('RATING')
plt.show()
```

```python
cond2=(data_revenue['Genre']=='Action,Adventure,Sci-Fi')
df2=data_revenue.loc[cond2].nlargest(5,'Revenue (Millions)')
df2.plot(kind='bar',x='Title',y='Revenue (Millions)',alpha=0.6,figsize=(14,8),title='Movies with Highest Revenues in Genre(Action,Adventure,Sci-Fi)',color='red')
plt.xlabel('TITLE')
plt.ylabel('REVENUE')
plt.show()
```

```python
cond3=(data_rating['Genre']=='Action,Adventure,Sci-Fi')
df3=data_rating.loc[cond3].nlargest(5,'Metascore')
df3.plot(kind='line',x='Title',y='Metascore',figsize=(14,8),title='Movies with Highest Metascore in Genre(Action,Adventure,Sci-Fi)',stacked=False,alpha=0.5,color='black',ylim=(70,100))
plt.xlabel('TITLE')
plt.ylabel('METTASCORE')
plt.show()
```

```python
df5=data_revenue.groupby('Year')['Revenue (Millions)'].max()
print(df5)
plt.figure(figsize=(10,8))
df5.plot(kind='bar',ylim=(0,1000),color='orange')
plt.title('Maximum Earning of each year')
plt.xlabel('Title')
plt.ylabel('Revenue (Millions)')
```

```python
df6=data_rating.groupby('Year')['Rating'].mean()
print(df6)
plt.figure(figsize=(12,8))
df6.plot(kind='area',xlim=(2005,2017),ylim=(6.0,8.0),color='green',alpha=0.7)
plt.title('Average Rating Of Each Year')
plt.xlabel('Year')
plt.ylabel('Rating')
```

```python
df7=data_revenue.groupby('Year')['Revenue (Millions)'].mean()
print(df7)
plt.figure(figsize=(10,8))
df7.plot(kind='bar',ylim=(0,130),color='blue')
plt.title('Average Earning of each year')
plt.xlabel('Title')
plt.ylabel('Revenue (Millions)')
```

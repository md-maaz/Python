```python
# Parsing of Date 

df5=pd.read_csv('melb_data.csv',usecols=["Suburb", "Address", "Date", "Type", "Regionname",  
              "Distance", "Price"],parse_dates=['Date'])
df5.head()
```

```python
df5.dtypes
```

```python
# check if there are any missing values 
df5.isnull().sum()
```

```python
# any other way to deal with missing data 
# drop the value/rows 
df5.dropna(axis=0,how='any')
# to make the change permanent use inplace 
df5.dropna(axis=0,how='any',inplace=True)
df5.isna().sum()
```

```python
df5.dtypes
```

```python
# changing data type of a column 
df5['Price']=df5['Price'].astype('int')
df5['Price'].dtypes
```

```python
# creating new columns 
# OR deriving columns from Existing Column 
df5["Price(Millions)"]=df5['Price']/1000000
df5.head()
```

```python
#insert the new column at a specific index 
#first drop the new column 
df5.drop('Price(Millions)',inplace=True,axis=1)
# using insert func 
df5.insert(4,'Price(Millions)',df5['Price']/1000000)
#display the head 
df5.head()
```

```python
# Uniques values in a Column
df5['Regionname'].unique()
```

```python
df5['Regionname'].nunique()
```

```python
df5['Regionname'].value_counts()
```

```python
# Filtering data 
# no of houses that cost more that 1million 
cond=(df5['Price(Millions)']>1)
# for filtering use loc , iloc
df5.loc[cond]
```

```python
df5['Type'].unique()
```

```python
# multiple filter cond
cond1= ((df5['Price(Millions)']>1) & (df5['Type']=='u'))
df5.loc[cond1]
```

```python
df5['Regionname'].unique()
```

```python
# 3 condn filter 
cond2= ((df5['Price(Millions)']>1) & (df5['Type']=='u') & (df5['Regionname']=='Southern Metropolitan'))
df5.loc[cond2]
```

```python
df5['Suburb'].unique()
```

```python
df5['Suburb'].value_counts()
```

```python
# using isin() func to filter data 
suburb_area=['Richmond','Beaconsfield Upper','Aspendale Gardens']
df5[df5['Suburb'].isin(suburb_area)]
```

```python
# filtering with str accessor 
df6=df5[df5['Suburb'].str.startswith('D')]
df6
```

```python
# in df6 index is same as of df5 
# ie to change the index of df6 
df6.reset_index(drop=True)
```

```python
#another method of filtering .query() method 
df7= df5.query("'Price(Millions)'>1" and  "Type=='u'" and "Regionname=='Northern Victoria'")
df7
```

```python
dis_mean=df7['Distance'].mean()
```

```python
# selecting suburb with distance greater / less than mean distance 
df8=df7[df7['Distance']>dis_mean]
df8
```

```python
df9=df7[df7['Distance']<=dis_mean]
df9
```

```python
# string accessor with split func  
df5['Region']=df5['Regionname'].str.split(" ",expand=True)[0]
df5
```

```python
# date accessor 
df5['Month'],df5['Year']=df5['Date'].dt.month,df5['Date'].dt.year
df5
```

```python
# group-by function 
df5[['Region','Price(Millions)']].groupby('Region',as_index=False).mean()
```

```python
# sorting rows based on columns values 
df5.sort_values('Price',ascending=True)
```

```python
df5.sort_values('Distance',ascending=False)[['Price(Millions)','Suburb','Year']].head(20)
```

```python
# pandas Visualization 
df5['Distance'].plot(kind='hist', figsize=(5,8), title='Distance')
```
